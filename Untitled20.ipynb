```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt 
import seaborn as sns
```

**Data Loading**

```python
df = pd.read_csv('My_clean_test_data.csv')
print(df.head())
```

```python
df.dropna()
```

```python
missing_values_count = df.isnull().sum()
missing_values_count[0:10]
```

```python
columns_with_na_dropped =df.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
pd.DataFrame(df).fillna(0)
```

```python
df = df.reset_index()
```

```python
np.isnan(df.any()) #and gets False
np.isfinite(df.all()) #and gets True
```

```python
del df['ClosedDate']
del df['DateCreditPulled']
```

```python
del df['LoanOriginationDate']
del df['FirstRecordedCreditLine']
del df['ListingCreationDate']
```

```python
X = df.iloc[:, :-1].values
Y = df.iloc[:, 4].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)
```

```python
scalar= StandardScaler()
scalar.fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)  
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric='euclidean')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
print(classification_report(y_test, y_pred))
print (confusion_matrix(y_test, y_pred))

```
